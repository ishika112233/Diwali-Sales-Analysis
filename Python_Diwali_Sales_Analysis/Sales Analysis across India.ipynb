## Exploratory Data Analysis

## Gender

```python
dataFrame.columns
```

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = dataFrame) #this will create bar chart without the values - sns is seaborn

for bars in ax.containers: # to show values of each back we used this
    ax.bar_label(bars)
```

```python
dataFrame.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=True)

# groupby is same as SQL - here we are groupby on the basis of gender and total sum of Amount and sorting 
# the Amount value in ascending order
```

```python
sale_gender = dataFrame.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sale_gender)

#Plotting the above value in bar chart 
```

#### From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

## AGE

```python
dataFrame.columns
```

```python
ax = sns.countplot(data = dataFrame, x = 'Age Group', hue = 'Gender') # Hue define legends in the graph

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sale_age = dataFrame.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sale_age)

#Plotting the above value in bar chart 
```

#### From above graphs we can see that most of the buyers are of age group between 26-35 yrs female

## State

```python
# total number of orders from top 10 states --- head to return top 10values 

sales_state = dataFrame.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)}) #fixing the size of chart - width and length
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

```python
#total number of amount spent by top 10 states 

sales_state = dataFrame.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)}) #fixing the size of chart - width and length
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
```

#### From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively
